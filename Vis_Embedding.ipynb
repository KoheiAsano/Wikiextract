```python
#benchmarkの読み込み
import pandas as pd
benchdf = pd.read_csv('./benchmark.txt',
                        sep=',',
                        encoding='utf-8',
                        index_col=False,
                       )
```

```python
from gensim.models import word2vec
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import random

#fontの設定(on BoW)
font = {'family':'TakaoGothic'}
plt.rc('font', **font)

#embeddingで作成したモデルの読み込み
unique_w2v = word2vec.Word2Vec.load("data/forplot.model")
words = list(unique_w2v.wv.vocab)

#plotする単語リスト作成、下記はテキストのunique vocabを赤、bench markを黒でプロット
newvocabplot = [[word,"r"] for word in words]
benchplot = [[word,"k"] for word in benchdf["BaseForm"] if word in words]
plotwords = random.sample(newvocabplot, 20) + random.sample(benchplot, 20)


length = len(plotwords)
data = []
 
j = 0
while j < length:
    data.append(unique_w2v[plotwords[j][0]])
    j += 1
 
pca = PCA(n_components=2)
pca.fit(data)
data_pca= pca.transform(data)
 
length_data = len(data_pca)
plt.figure(figsize=(10,10),dpi=900)
i = 0
while i < length_data:
    #点プロット
    plt.plot(data_pca[i][0], data_pca[i][1], ms=5.0, zorder=2, marker="x", color=plotwords[i][1])
 
    #文字プロット
    plt.annotate(plotwords[i][0], (data_pca[i][0], data_pca[i][1]), size=7)
 
    i += 1

plt.show()
```
