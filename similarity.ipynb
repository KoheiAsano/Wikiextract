```python
from gensim.models import word2vec
unique_w2v = word2vec.Word2Vec.load("data/forplot.model")
```

```python
words = list(unique_w2v.wv.vocab)

import pandas as pd
benchdf = pd.read_csv('./benchmark.txt',
                        sep=',',
                        encoding='utf-8',
                        index_col=False,
                       )


newwords = [new for new in words if not new in benchdf["BaseForm"] ]
```

```python
for new in newwords:
    print(new)
    sim_words = [sim[0] for sim in unique_w2v.wv.most_similar(new)]
    print(f"類似語{sim_words}")
```
